```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from Clean import clean
from sklearn.linear_model import LogisticRegression
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score, GridSearchCV
from sklearn.metrics import f1_score,roc_auc_score,accuracy_score,confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

%matplotlib inline
```

```python
df = pd.read_json("data/data.json")
df.sample(20).T
```

```python
df.info()
```

```python
def clean(df):
    fraudster =  ['fraudster_event' ,'fraudster','fraudster_att']
    df["Target"] = [a in fraudster for a in df['acct_type'] ]
    df["Target"] = df["Target"].astype(int)
    return df
```

```python
df = clean(df)
df["Target"].value_counts()
```

```python
total_cost = [np.sum([b['cost']*b['quantity_total'] for b in a]) for a in df['ticket_types'] ]
df['total_cost'] = total_cost
df[["Target",'total_cost']].groupby('Target')['total_cost'].mean()
```

```python
sub= df[['name_length','num_payouts','user_age','org_facebook','org_twitter','body_length','gts','sale_duration','Target']]
sns.pairplot(data=sub,hue='Target',plot_kws = {'alpha': 0.1})
```

```python
from sklearn.ensemble import RandomForestClassifier
sub= df[['name_length','num_payouts','user_age','org_facebook','org_twitter','body_length','gts','sale_duration','Target']]

y = sub.iloc[:,-1]
X = sub.iloc[:,:-1]
X['org_facebook'] = X['org_facebook'].fillna(-1)
X['org_twitter'] = X['org_twitter'].fillna(-1)
X['sale_duration'] = X['sale_duration'].fillna(-1)



X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)
rfc = RandomForestClassifier()

cross_val_score(rfc,X_train,y_train,cv=5,scoring="roc_auc")
```

```python
rfc.fit(X_train,y_train)
y_test_predict = rfc.predict(X_test)
roc_auc_score(y_test,y_test_predict)
```
